## Commonsense qa

````python
# ── Install these once in your environment: ────────────────────────────────────
# pip install transformers datasets sentence-transformers

import torch
import numpy as np
from transformers import AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import SentenceTransformer
from datasets import load_dataset

class Se2ICLInference:
    def __init__(
        self,
        model_name: str = "google/gemma-2b",      # switched to Gemma-2B
        embed_model: str = "all-MiniLM-L6-v2",
        shot: int = 3,
        retrieve_k: int = 20,
        beam_size: int = 3,
        device: str = None,
    ):
        # device
        self.device = device or ("cuda" if torch.cuda.is_available() else "cpu")

        # 1) bi-encoder for retrieval
        self.embedder = SentenceTransformer(embed_model, device=self.device)

        # 2) Gemma-2B for CE scoring
        self.tok = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
        if self.tok.pad_token is None:
            self.tok.pad_token = self.tok.eos_token
        self.lm = AutoModelForCausalLM.from_pretrained(
            model_name,
            trust_remote_code=True,
            torch_dtype=torch.float16,
            device_map="auto"
        ).to(self.device)
        self.lm.eval()

        self.shot       = shot
        self.retrieve_k = retrieve_k
        self.beam_size  = beam_size

    def _batch_ce(self, texts):
        """Return total cross-entropy for each string in texts."""
        enc = self.tok(
            texts,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=512
        ).to(self.device)
        with torch.no_grad():
            out = self.lm(enc.input_ids, labels=enc.input_ids)
        lengths = enc.attention_mask.sum(dim=1).float()
        return (out.loss * lengths).cpu().tolist()

    def beam_search(self, prompt: str, gold: str):
        """
        1) retrieve top-K demos via SBERT cosine
        2) beam-search over those K with CE-rerank
        """
        sims, idxs = (self.demos_emb @ self.q_emb.T).squeeze(1).topk(self.retrieve_k)
        candidates = [self.demos[i] for i in idxs.cpu().tolist()]

        beams = [([], 0.0)]
        for _ in range(self.shot):
            new_beams, xb, cb, parents = [], [], [], []
            for seq, cum in beams:
                for d in candidates:
                    if d in seq:
                        continue
                    xb.append(f"{d}\n\n{self.current_prompt}\nAnswer: {gold}")
                    cb.append(f"{d}\n\nAnswer: {gold}")
                    parents.append((seq, d, cum))
            H_xb = self._batch_ce(xb)
            H_cb = self._batch_ce(cb)
            for (seq, d, cum), hxb, hcb in zip(parents, H_xb, H_cb):
                new_beams.append((seq + [d], cum + (hxb - hcb)))
            beams = sorted(new_beams, key=lambda x: x[1], reverse=True)[:self.beam_size]
        return beams[0][0]

    def classify_by_ce(self, context: str, choices: list[str]) -> str:
        """Pick the choice with lowest CE."""
        texts = [f"{context}\n\nAnswer: {c}" for c in choices]
        ces   = self._batch_ce(texts)
        return choices[int(np.argmin(ces))]

    def evaluate(self, prompts, choice_lists, labels):
        # precompute demo-pool
        self.demos      = [f"{p}\nAnswer: {l}" for p, l in zip(prompts, labels)]
        self.demos_emb  = self.embedder.encode(self.demos, convert_to_tensor=True)

        correct = 0
        for i, (p, choices, gold) in enumerate(zip(prompts, choice_lists, labels)):
            # prepare prompt embedding & store for beam_search
            self.current_prompt = p
            self.q_emb          = self.embedder.encode([p], convert_to_tensor=True)

            # 1) pick demos
            demos = self.beam_search(p, gold)

            # 2) classify
            ctx  = "\n\n".join(demos) + f"\n\n{p}"
            pred = self.classify_by_ce(ctx, choices)

            ok = (pred.lower() == gold.lower())
            correct += ok

            if i < 3:
                print(f"\n=== Example {i} ===")
                print("Prompt:", p.replace("\n"," ↵ "))
                print("Choices:", choices)
                print("Selected demos:")
                for d in demos:
                    print(" •", d.replace("\n"," ↵ "))
                print(f"Predicted: {pred!r}, Gold: {gold!r} →", "✓" if ok else "✗")

        acc = correct / len(prompts)
        print(f"\n⇒ Accuracy: {100*acc:.2f}% ({correct}/{len(prompts)})")


if __name__=="__main__":
    # 1) load a small slice of CommonsenseQA
    ds = load_dataset("tau/commonsense_qa", split="train[:200]").shuffle(seed=0)

    # 2) build prompts, choices, labels
    prompts, choice_lists, labels = [], [], []
    for ex in ds:
        q    = ex["question"]
        lab, txt = ex["choices"]["label"], ex["choices"]["text"]
        prompt   = "Question: " + q + "\n" + "\n".join(f"{l}. {t}" for l,t in zip(lab, txt))
        prompts.append(prompt)
        choice_lists.append(txt)
        labels.append(txt[lab.index(ex["answerKey"])])

    # 3) run with Gemma-2B
    selector = Se2ICLInference(
        model_name   = "google/gemma-2b",
        embed_model  = "all-MiniLM-L6-v2",
        shot         = 3,
        retrieve_k   = 20,
        beam_size    = 3,
    )
    selector.evaluate(prompts, choice_lists, labels)
````

## AG News

````python
# ── Install these once in your environment: ────────────────────────────────────
# pip install transformers datasets sentence-transformers

import torch
import numpy as np
from transformers import AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import SentenceTransformer
from datasets import load_dataset

class Se2ICLAGNews:
    def __init__(
        self,
        model_name: str = "google/gemma-2b",
        embed_model: str = "all-MiniLM-L6-v2",
        shot: int = 3,
        retrieve_k: int = 20,
        beam_size: int = 3,
        device: str = None,
    ):
        # device
        self.device = device or ("cuda" if torch.cuda.is_available() else "cpu")

        # 1) SBERT for fast retrieval
        self.embedder = SentenceTransformer(embed_model, device=self.device)

        # 2) Gemma-2B for CE scoring
        self.tok = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
        if self.tok.pad_token is None:
            self.tok.pad_token = self.tok.eos_token
        self.lm = AutoModelForCausalLM.from_pretrained(
            model_name,
            trust_remote_code=True,
            torch_dtype=torch.float16,
            device_map="auto"
        ).to(self.device)
        self.lm.eval()

        self.shot       = shot
        self.retrieve_k = retrieve_k
        self.beam_size  = beam_size

        # fixed AG-News categories
        self.choices = ["World", "Sports", "Business", "Sci/Tech"]

    @torch.inference_mode()
    def _batch_ce(self, texts, batch_size=8):
        """Return total cross-entropy for each string in texts."""
        ces = []
        for i in range(0, len(texts), batch_size):
            chunk = texts[i : i+batch_size]
            enc = self.tok(
                chunk,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=512
            ).to(self.device)
            out = self.lm(enc.input_ids, labels=enc.input_ids)
            lengths = enc.attention_mask.sum(dim=1).float()
            ces.extend((out.loss * lengths).cpu().tolist())
        return ces

    def beam_search(self, gold: str):
        """
        1) retrieve top-K demos via SBERT cosine
        2) perform shot-by-shot beam-search with CE difference
        """
        sims, idxs = (self.demos_emb @ self.q_emb.T).squeeze(1).topk(self.retrieve_k)
        candidates = [self.demos[i] for i in idxs.cpu().tolist()]

        beams = [([], 0.0)]
        for _ in range(self.shot):
            new_beams, xb, cb, parents = [], [], [], []
            for seq, cum in beams:
                for d in candidates:
                    if d in seq:
                        continue
                    # CE with and without appending this demo
                    xb.append(f"{d}\n\n{self.current_prompt}\nAnswer: {gold}")
                    cb.append(f"{d}\n\nAnswer: {gold}")
                    parents.append((seq, d, cum))
            H_xb = self._batch_ce(xb)
            H_cb = self._batch_ce(cb)
            for (seq, d, cum), hxb, hcb in zip(parents, H_xb, H_cb):
                new_beams.append((seq + [d], cum + (hxb - hcb)))
            beams = sorted(new_beams, key=lambda x: x[1], reverse=True)[: self.beam_size]
        return beams[0][0]

    def classify_by_ce(self, context: str) -> str:
        """Pick the category with lowest CE under the model."""
        texts = [f"{context}\n\nAnswer: {c}" for c in self.choices]
        ces   = self._batch_ce(texts)
        return self.choices[int(np.argmin(ces))]

    def evaluate(self, prompts, labels):
        # 1) build and embed the demo-pool once
        self.demos     = [f"{p}\nAnswer: {l}" for p,l in zip(prompts, labels)]
        self.demos_emb = self.embedder.encode(self.demos, convert_to_tensor=True)

        correct = 0
        for i, (p, gold) in enumerate(zip(prompts, labels)):
            # store prompt for beam_search
            self.current_prompt = p
            self.q_emb          = self.embedder.encode([p], convert_to_tensor=True)

            # 2) select demos
            demos = self.beam_search(gold)

            # 3) classify by CE
            ctx  = "\n\n".join(demos) + f"\n\n{p}"
            pred = self.classify_by_ce(ctx)

            if pred == gold:
                correct += 1

            if i < 3:  # debug print first 3
                print(f"\n=== Example {i} ===")
                print("Prompt:", p.replace("\n"," ↵ "))
                print("Selected demos:")
                for d in demos:
                    print(" •", d.replace("\n"," ↵ "))
                print(f"Predicted: {pred!r}, Gold: {gold!r}")

        acc = correct / len(prompts)
        print(f"\n⇒ AG-News Accuracy: {100*acc:.2f}% ({correct}/{len(prompts)})")


if __name__=="__main__":
    # 1) load a small slice of AG-News
    ds = load_dataset("ag_news", split="train[:200]").shuffle(seed=0)

    # 2) build prompts & labels
    prompts = [
        f"News article:\n{ex['text']}\n\nCategory? (World / Sports / Business / Sci/Tech)"
        for ex in ds
    ]
    labels = ["World", "Sports", "Business", "Sci/Tech"]
    labels = [labels[ex["label"]] for ex in ds]

    # 3) run
    selector = Se2ICLAGNews(
        model_name   = "google/gemma-2b",
        embed_model  = "all-MiniLM-L6-v2",
        shot         = 3,
        retrieve_k   = 20,
        beam_size    = 3,
    )
    selector.evaluate(prompts, labels)
````

````python
# ── Install these once in your environment: ────────────────────────────────────
# pip install transformers datasets sentence-transformers accelerate

import torch
import numpy as np
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import SentenceTransformer

class Se2ICLInference:
    def __init__(
        self,
        model_name: str = "google/gemma-2b",      # switched to Gemma-2B
        embed_model: str = "all-MiniLM-L6-v2",
        shot: int = 3,
        retrieve_k: int = 20,
        beam_size: int = 3,
        device: str = None,
    ):
        # device
        self.device = device or ("cuda" if torch.cuda.is_available() else "cpu")

        # 1) Bi-encoder for fast retrieval
        self.embedder = SentenceTransformer(embed_model, device=self.device)

        # 2) Causal LM for CE scoring
        self.tok = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
        if self.tok.pad_token is None:
            self.tok.pad_token = self.tok.eos_token
        self.lm = AutoModelForCausalLM.from_pretrained(
            model_name,
            trust_remote_code=True,
            torch_dtype=torch.float16,
            device_map="auto"
        ).to(self.device)
        self.lm.eval()

        self.shot       = shot
        self.retrieve_k = retrieve_k
        self.beam_size  = beam_size

        # fixed SST-5 choices
        self.choices = [
            "Very Negative",
            "Negative",
            "Neutral",
            "Positive",
            "Very Positive"
        ]

    def _batch_ce(self, texts):
        """Compute total cross-entropy for each input in `texts`."""
        enc = self.tok(
            texts,
            return_tensors="pt",
            padding=True,
            truncation=True,
            max_length=512
        ).to(self.device)
        with torch.no_grad():
            out = self.lm(enc.input_ids, labels=enc.input_ids)
        lengths = enc.attention_mask.sum(dim=1).float()
        return (out.loss * lengths).cpu().tolist()

    def beam_search(self, prompt: str, gold: str):
        """Pick `shot` demos via SBERT retrieval + beam-search CE rerank."""
        # 1) top-K retrieval
        q_emb      = self.embedder.encode([prompt], convert_to_tensor=True)
        sims, idxs = (self.demos_emb @ q_emb.T).squeeze(1).topk(self.retrieve_k)
        cands      = [self.demos[i] for i in idxs.cpu().tolist()]

        # 2) beam-search
        beams = [([], 0.0)]
        for _ in range(self.shot):
            new, xb, cb, parents = [], [], [], []
            for seq, score in beams:
                for d in cands:
                    if d in seq:
                        continue
                    xb.append(f"{d}\n\n{prompt}\nAnswer: {gold}")
                    cb.append(f"{d}\n\nAnswer: {gold}")
                    parents.append((seq, d, score))
            H_xb = self._batch_ce(xb)
            H_cb = self._batch_ce(cb)
            for (seq, d, score), hxb, hcb in zip(parents, H_xb, H_cb):
                new.append((seq + [d], score + (hxb - hcb)))
            beams = sorted(new, key=lambda x: x[1], reverse=True)[: self.beam_size]

        return beams[0][0]

    def classify_by_ce(self, context: str) -> str:
        """
        Given context (demos + prompt), append each fixed choice,
        score by CE, and return the choice with lowest CE.
        """
        texts = [f"{context}\n\nAnswer: {c}" for c in self.choices]
        ces   = self._batch_ce(texts)
        return self.choices[int(np.argmin(ces))]

    def evaluate(self, prompts, labels):
        # build demo-pool
        self.demos     = [f"{p}\nAnswer: {l}" for p,l in zip(prompts, labels)]
        self.demos_emb = self.embedder.encode(self.demos, convert_to_tensor=True)

        correct = 0
        for i, (p, gold) in enumerate(zip(prompts, labels)):
            # 1) pick demos
            demos = self.beam_search(p, gold)

            # 2) assemble context
            ctx = "\n\n".join(demos) + f"\n\n{p}"

            # 3) classify
            pred = self.classify_by_ce(ctx)

            if pred == gold:
                correct += 1

            if i < 3:
                print(f"\n=== Example {i} ===")
                print("Review prompt:", p.replace("\n"," ↵ "))
                print("Selected demos:")
                for d in demos:
                    print(" •", d.replace("\n"," ↵ "))
                print(f"Predicted: {pred!r}, Gold: {gold!r} →", "✓" if pred==gold else "✗")

        acc = correct / len(prompts)
        print(f"\n⇒ SST-5 Accuracy: {100*acc:.2f}% ({correct}/{len(prompts)})")


if __name__=="__main__":
    # 1) load a small slice of SST-5
    ds = load_dataset("SetFit/sst5", split="train[:200]").shuffle(seed=0)

    # 2) build prompts & labels
    label_map = {
        0: "Very Negative",
        1: "Negative",
        2: "Neutral",
        3: "Positive",
        4: "Very Positive"
    }
    prompts, labels = [], []
    for ex in ds:
        review = ex["text"]
        p = (
            f"Review: {review}\n\n"
            "Sentiment? (Very Negative / Negative / Neutral / Positive / Very Positive)"
        )
        prompts.append(p)
        labels.append(label_map[ex["label"]])

    # 3) instantiate & evaluate with Gemma-2B
    selector = Se2ICLInference(
        model_name   = "google/gemma-2b",
        embed_model  = "all-MiniLM-L6-v2",
        shot         = 3,
        retrieve_k   = 20,
        beam_size    = 3,
    )
    selector.evaluate(prompts, labels)
````
